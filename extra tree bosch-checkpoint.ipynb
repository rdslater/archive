```python
import pandas as pd
import numpy as np
import re
train = pd.read_csv("./kaggle/Bosch/data/train_numeric.csv", chunksize=100000)
train_new=pd.DataFrame()
toggle=0
for chunk in train:
    print("-",end='')
    if toggle == 0:
        a=chunk.columns.values.tolist()
        myString = '(L[0-9]*_S[0-9]*_) *'
        pattern1 = r'\b' + re.escape(myString) + r'\b'
        pattern=re.compile(myString)
        indices = [i for i, x in enumerate(a) if re.search(pattern, x)]
        i=0
        for item in a:
            if 'L' in item:
                a[i]=re.search(pattern, a[i]).group(1)
            i=i+1
        a=list(set(a))
    new_train=pd.DataFrame()
    for item in a:
        if (item=='Id' )| (item=='Response'):
            new_train[item]=chunk[item]
        else:
            spike_cols = [col for col in chunk.columns if item in col]
            mean=item+'mean'
            nacount=item+'NAcount'
            max_v=item+'max'
            min_v=item+'min'
            new_train[mean]=chunk[spike_cols].mean(axis=1)
            new_train[nacount]=chunk[spike_cols].isnull().sum(axis=1)
            new_train[max_v]=chunk[spike_cols].max(axis=1)
            new_train[min_v]=chunk[spike_cols].min(axis=1)
            cols=new_train.columns.values.tolist()
            cols.sort()
            new_train=new_train[cols]
        
    if toggle==0:
        new_train.to_csv("./kaggle/Bosch//data/train_num_smry.csv",mode='w',index=False,header=True)
        toggle=1
    else:
        new_train.to_csv( "./kaggle/Bosch//data/train_num_smry.csv",mode='a',index=False,header=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
train=pd.read_csv("./kaggle/Bosch//data/train_numeric.csv",nrows=200000)

extc = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=4, random_state=0, criterion='entropy',verbose=1,n_jobs=-1)

target=train['Response']
ids=train['Id']
train=train.drop(['Response','Id'],axis=1)
train=train.fillna(value=-100)
extc.fit(train,target)
import matplotlib.pyplot as plt
#clf4 = ExtraTreesClassifier(n_estimators=20, max_depth=None, min_samples_split=4, random_state=0, criterion='entropy',verbose=1,n_jobs=-1)
#extc.fit(train,target)

#print(scores0)
importances = extc.feature_importances_
std = np.std([tree.feature_importances_ for tree in extc.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
cols=list(train.columns.values)

print("Feature ranking:")
feature_names=[]
for f in range(train.shape[1]):
    print("%d. %s (%f) (%f) (%s)" % (f + 1, cols[indices[f]], importances[indices[f]], std[indices[f]], train.iloc[:,indices[f]].dtype))
    feature_names.append([cols[indices[f]], std[indices[f]]-importances[indices[f]]])
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(train.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(train.shape[1]), indices)
plt.xlim([-1, train.shape[1]])
plt.show()
```

```python

import time

from sklearn.cross_validation import KFold
import itertools
import pandas as pd
import numpy as np
import re
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
%matplotlib inline

train=pd.read_csv("./kaggle/Bosch//data/train_numeric.csv", chunksize=50000)
train2=pd.read_csv("./kaggle/Bosch/data/train_cat_factor.csv", chunksize=50000)


start = time.time()
m=25
n=9

clf6 =GradientBoostingClassifier(n_estimators=m, max_depth=n, min_samples_split=4, random_state=0,warm_start=True)
response=pd.DataFrame()

for (chunk_num,chunk_cat) in  itertools.zip_longest(train,train2):
    response=chunk_num['Response']
    chunk_num.drop(['Id','Response'],axis=1,inplace=True)
    chunk_num=chunk_num.fillna(-100)
    chunk_cat=chunk_cat.fillna(value=0)
    chunk_cat.drop(['Id'],axis=1,inplace=True)
    chunk=pd.concat([chunk_num,chunk_cat],axis=1)
    chunk=np.ascontiguousarray(chunk)
    kf = KFold( chunk.shape[0],n_folds=2,shuffle=True)
    for train_id,test_id in kf:
        X_train=chunk[train_id]
        Y_train=response[train_id]
        X_test=chunk[test_id]
        Y_test=response[test_id]
    clf6.fit(X_train,Y_train)
    
    thresholds = np.linspace(0.01, 0.99, 50)
    y=Y_test
        
    predict=clf6.predict_proba(X_test)
    preds=predict[:,1]
    mcc = np.array([matthews_corrcoef(y, preds>thr) for thr in thresholds])
    plt.plot(thresholds, mcc)
    best_threshold = thresholds[mcc.argmax()]
    print(mcc.max(),n)
    m=m+25
    clf6.n_estimators=m
    end = time.time()
    print(end-start)
    
```
