```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ggplot import *
del data1
path='c://Users//Robert//Downloads//nyt1.csv'
data1=pd.DataFrame(pd.read_csv(path))
data1['age_cat']='a'
data1['age_cat']=np.where(data1['Age']==0,'0',data1['age_cat'])
data1['age_cat']=np.where(data1['Age']<0,'<0',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>0,data1['Age']<=18),'1-18',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>18,data1['Age']<=25),'19-25',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>25,data1['Age']<=35),'26-35',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>35,data1['Age']<=45),'36-45',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>45,data1['Age']<=55),'46-55',data1['age_cat'])
data1['age_cat']=np.where(np.logical_and(data1['Age']>55,data1['Age']<=65),'56-65',data1['age_cat'])
data1['age_cat']=np.where(data1['Age']>65,'65+',data1['age_cat'])

```

```python
data2=data1[data1['Impressions']>0]
data2['th']=data2['Clicks']/data2['Impressions']

data2
```

```python
ggplot(data2,aes(x='th',colour='age_cat')) + geom_density()
```
