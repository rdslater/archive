```python
import pandas as pd
import numpy as np
train = pd.read_hdf('train.h5')
```

```python
cols=train.columns.tolist()
for label in cols:
    med=train[label].median()
    train[label]=train[label].fillna(med)
    
```

```python
test=train[train['id']>=960]
train=train[train['id']<960]
```

```python
time=train.timestamp
x_train=train.drop(['y','timestamp'],axis=1)
y_train=train.y
x_test=test.drop(['y'],axis=1)
y_test=test.y
from sklearn.linear_model import Ridge
gml=Ridge(alpha=1.0, fit_intercept=False, normalize=False, max_iter=None, tol=0.001, solver='auto', random_state=None)
```

```python
gml.fit(x_train,y_train)
```

results=pd.DataFrame({'label':cols[:-1],'coef':gml.coef_})
